# Loan Prediction Data Analysis


```python
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the data
df = pd.read_csv('D:/Python Practice Data/Project/Loan Prediction.csv')  # Make sure to place your file in the same directory

# Step 3: Overview of data
print("Data Sample:")
print(df.head())
print("\nColumn Names:")
print(df.columns)
df.isnull().sum()
```

```python
cat_cols = ['Married/Single', 'House_Ownership', 'Car_Ownership', 'Profession', 'CITY', 'STATE']

for col in cat_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

```python
# step2: print("\nBasic Info:")
print(df.info())
```

```python
# Step 3: Basic Info and Preview
print("Dataset Preview:")
display(df.head())
```

```python
# Step 4: Total Customers
total_customers = df.shape[0]
print(f"\nTotal Customers: {total_customers}")
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Step 5: Age Analysis
average_age = df['Age'].mean()
print(f"\nAverage Age: {average_age:.2f} years")

plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Step 6: Income Analysis
average_income = df['Income'].mean()
print(f"\nAverage Income: {average_income:,.0f}")

plt.figure(figsize=(8, 5))
sns.histplot(df['Income'], bins=30, kde=True, color='lightgreen')
plt.title('Income Distribution')
plt.xlabel('Income (INR)')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Step 7: Risk Analysis
risk_counts = df['Risk_Flag'].value_counts()

plt.figure(figsize=(4, 2))
sns.barplot(x=risk_counts.index, y=risk_counts.values, palette=['lightgreen', 'lightcoral'])
plt.title('Risk Flag Distribution')
plt.xlabel('Risk Flag (0 = Low Risk, 1 = High Risk)')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()
```

```python
# Step 8: Age vs Income (Optional Analysis)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Income', hue='Risk_Flag', data=df, palette=['green', 'red'])
plt.title('Age vs Income (Colored by Risk Flag)')
plt.xlabel('Age')
plt.ylabel('Income (INR)')
plt.grid(True)
plt.show()

# Done! You can add more analysis (like marital status breakdown, profession analysis, etc.)
```

1 Age vs Loan Risk
The majority of high-risk customers (Risk_Flag = 1) are concentrated in the younger age group (21-35 years).
Older customers (50+ years) generally show lower risk, possibly due to stable income and job history.

2 Income Disparity
Income varies drastically, from as low as 10,310 to nearly 10 million.
However, customers with extremely low incomes (under 500,000) are more likely to be high-risk.
This indicates that low-income customers face greater difficulty in loan repayment, which aligns with real-world patterns.

3 Marital Status Impact
Preliminary exploration (if added to the analysis) might show:
Married customers tend to have more stable risk profiles compared to singles.
This could be because married customers often have more shared financial responsibilities and incomes.

4 House Ownership Stability
Customers who own houses tend to have lower risk scores, compared to those living in rented houses.
This is a strong signal that home ownership correlates with financial stability.

5 Job Experience vs Risk
Customers with less than 3 years of job experience show a higher proportion of high-risk flags.
This suggests that newer employees are riskier for loans, possibly due to unstable employment or lower savings.

6 High-Income Doesnt Always Mean Low Risk
Surprisingly, even some customers with incomes above 7 million were marked as high-risk.
This highlights that income alone isnt a perfect predictor  factors like spending habits, debts,
and financial discipline also matter.


 Business Implication


 For loan approval models, it's important to consider a combination of factors  age,
   income, house ownership, and job stability  rather than relying on income alone.


 Targeted loan products could be created for younger, lower-income customers, but with
    stricter repayment terms or smaller loan amounts. 

 Encourage financial literacy programs for young customers to improve their long-term loan eligibility.

```python
# 4. Risk by Car Ownership
plt.figure(figsize=(4,3))
sns.countplot(x='Car_Ownership', hue='Risk_Flag', data=df, palette='Set1')
plt.title('Risk by Car Ownership')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# --- Customer Attraction Insights ---

# 6. Top 10 Professions Applying for Loans
plt.figure(figsize=(8,4))
top_professions = df['Profession'].value_counts().head(10)
sns.barplot(y=top_professions.index, x=top_professions.values, palette='viridis')
plt.title('Top 10 Professions Applying for Loans')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 7. Income Distribution (who applies for loans)
plt.figure(figsize=(6,3))
sns.histplot(df['Income'], bins=20, kde=True, color='skyblue')
plt.title('Income Distribution of Applicants')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 8. Top 10 Cities with Highest Loan Applicants
plt.figure(figsize=(8,4))
top_cities = df['CITY'].value_counts().head(10)
sns.barplot(y=top_cities.index, x=top_cities.values, palette='magma')
plt.title('Top 10 Cities with Highest Loan Applicants')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
